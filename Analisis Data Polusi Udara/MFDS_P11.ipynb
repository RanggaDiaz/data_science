```python
import pandas as pd
import numpy as np
import string
import re
```

```python
df = pd.read_csv('polusiudarajakarta_data.csv', sep=';')
df.head()
```

```python
tweet = df['full_text']
```

```python
def clean_text(text):
    # Menghapus Username
    text = re.sub('@[^\s]+', '', text)
    # Menghapus URL
    text = re.sub(r'https?://\S+', '', text)
    # Menghapus tanda baca
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Menghapus karakter khusus atau tanda baca
    text = re.sub(r'[^\w\s]', '', text)
    # Menghapus karakter HTML
    text = re.sub(r'<.*?>', '', text)
    # Mempertahankan huruf dan angka
    text = re.sub('[^a-zA-Z0-9]', ' ', text)
    # Mengganti line baru dengan spasi
    text = re.sub('\n', ' ', text)
    # Mengubah ke huruf kecil
    text = text.lower()
    # Menghapus single character
    text = re.sub(r'\b\w\b', '', text)
    # Memisahkan dan menggabungkan kata
    text = ' '.join(text.split())
    return text
```

```python
df['clean'] = [clean_text(i) if isinstance(i, str) else '' for i in df['full_text']]
df.head(10)
```

```python
from nltk.tokenize import RegexpTokenizer
regexp = RegexpTokenizer(r'\w+|$[0-9]+|\S+')
df['Token'] =df ['clean'].apply(regexp.tokenize)
```

```python
normalized_word = pd.read_csv("/content/normalisasi.csv", encoding='latin1')

normalized_word_dict={}
for index, row in normalized_word.iterrows():
  if row[0] not in normalized_word_dict:
    normalized_word_dict[row[0]] = row [1]

def normalized_term(document):
  return [normalized_word_dict[term] if term in normalized_word_dict else term for term in document]

df['Normalisasi'] = df['Token'].apply(normalized_term)
df.head(10)
```

```python
import nltk
nltk.download('stopwords')  # Download the stopwords resource

from nltk.corpus import stopwords
stopword = stopwords.words('indonesian')

# Assuming you have a file named "stopwords.txt" with custom stopwords
import pandas as pd

txt_stopword = pd.read_csv("/content/stopwords.txt", names=["stopwords"], header=None)
stopword.extend(['yang', 'di', 'dan', 'yg', 'itu', 'dari', 'ke', 'untuk', 'yg', 'juga', 'ada', 'sumbernya', 'tetap', 'dalam', 'pada', 'ini', 'tersebut', 'semua', 'dengan', 'oleh', 'sudah', 'akan', 'bs', 'bsi', 'krn', 'hrs', 'pdhal', 'entah', 'apa', 'hasil', 'akhir', 'ingin', 'capai', 'gak', 'aja', 'krn', 'tiap', 'bulan', 'sprti', 'apa', 'mau', 'dr', 'keadaan', 'yg', 'bs', 'menjadi', 'bahan', 'pertimbangan', 'unt', 'penataan', 'ulang', 'semua', 'hal', 'salah', 'sdh', 'terjadi', 'selama', 'thn', 'ya', 'guys', 'seluruh', 'indonesia', 'hrs', 'kuatir', 'terhadap', 'masalah', 'polusi'])
stopword.extend(txt_stopword["stopwords"][0].split(' '))
stopword = set(stopword)

def remove_stopwords(text):
    text = [word for word in text if word not in stopword]
    return text

# Assuming you have a DataFrame named "df" with a column named "Normalisasi"
df['Stopwords'] = df['Normalisasi'].apply(remove_stopwords)
df
```
